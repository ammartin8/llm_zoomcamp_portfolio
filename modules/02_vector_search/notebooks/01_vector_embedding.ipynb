## Vector Embedding

````python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
````

````python
# Turning the query into a vector
q1 = "Can I still join the course after the start date?"
v1 = model.encode(q1)
v1
````

````python
len(v1)
````

v1 is a vector, an array of 384 numbers. Each number stands for some concept the model learned. We can't read off what any one of them means. But two vectors with similar values point to texts about similar things.

````python
# Encoding the document
d  = "You don't need to register. You're accepted. You can also just start learning and submitting homework without registering."
dv = model.encode(d)
dv
````

````python
len(dv)
````

````python
# Next, we compare the query against the document using dot product:
v1.dot(dv)
````

````python
# Trying an unrelated query
q2 = "How to install Docker on Windows?"
v2 = model.encode(q2)
v2
````

````python
v2.dot(dv)
````

The first query (asking about registration) (.32) is higher, therefore this query is more similar to the document about registration compared to second query about docker (.01). The dot product tells us how similar the document is to the query in question.

## Embedding a Dataset 

````python
# Loading data source
from ingest import load_faq_data

documents = load_faq_data()
````

````python
# question and answer will be embedded both together so that a query can match against both the question text and answer text in the index.
texts = []

for doc in documents:
    text = doc["question"] + " " + doc["answer"]
    texts.append(text)
````

````python
# Next step is to generate embeddings
# We need to embed in batches as processing can take a long time
# import tqdm to review process
from tqdm.auto import tqdm
````

````python
# Now chunk the dataset into batches and encode each batch
batch_size = 50
vectors = []

for i in tqdm(range(0, len(texts), batch_size)):
    batch = texts[i:i + batch_size]
    batch_vectors = model.encode(batch)
    vectors.extend(batch_vectors)

len(vectors)
````

````python
# Next we turn it into a 2-D array (matrix) where row = docs and cols = dimension of the vector which represent clues or meanings of the text the model has learned and is represented by a number
import numpy as np
X = np.array(vectors)
X
````

````python
len(X)
````

## Scoring the Documents

````python
# First we embed the query
query = "Can I still join the course after the start date?"
v_query = model.encode(query)
````

````python
print(f"Shape: {v_query.shape},\n View of array: {v_query}")
````

````python
# Compute dot product against all docs
scores = X.dot(v_query)
print(f"Shape:{scores.shape} \n\n View of output: {scores}")
````

Could be done with for loop but X.dot(v_query) is much faster. This results in one score per document.
```
scores = [v_query.dot(X[i]) for i in range(len(X))]
```

````python
# Finding the best match; The highest score is the most similar document
idx = np.argmax(scores)
idx, scores[idx]
````

This returns document 2 with score 0.76.

````python
# reviewing document
documents[idx]
````

## Pulling Top 5 Results
np.argsort sorts from lowest to highest, so the last 5 are the top ones:

````python
top5 = np.argsort(scores)[-5:]
top5
````

````python
# Next we reverse the order so it's highest to lowest in terms of similarity
top5 = top5[::-1]
top5
````

````python
scores[top5]
````

````python
#  Alternative: We negate the scores first, so the largest becomes the smallest. Then argsort puts the best matches at the front.
top5 = np.argsort(-scores)[:5]
top5
````

````python
# Reviewing the top 5 docs
for idx in top5:
    print(scores[idx])
    print(documents[idx])
    print()
````

This is vector search in its simplest form. We embed the query, compute dot products against all documents, and return the highest-scoring ones.
